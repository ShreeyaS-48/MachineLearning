```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df_train = pd.read_csv('/kaggle/input/fashionmnist/fashion-mnist_train.csv')
df_train.head()
```

```python
df_test = pd.read_csv('/kaggle/input/fashionmnist/fashion-mnist_test.csv')
df_test.head()
```

```python
X_train = df_train.iloc[:, 1:]
X_test = df_test.iloc[:, 1:]
y_test = df_test.iloc[:, 0]
y_train = df_train.iloc[:, 0]
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train_np = X_train.values
X_test_np  = X_test.values
X_train_np = X_train_np.reshape(-1, 28, 28, 1)
X_test_np  = X_test_np.reshape(-1, 28, 28, 1)
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Conv2D, AveragePooling2D, Flatten
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential()

model.add(Conv2D(6, kernel_size=(5,5), padding='valid', activation='tanh', input_shape = (28, 28, 1)))
model.add(AveragePooling2D(pool_size=(2,2), strides=2, padding='valid'))

model.add(Conv2D(16, kernel_size=(5,5), padding='valid', activation='tanh'))
model.add(AveragePooling2D(pool_size=(2,2), strides=2, padding='valid'))

model.add(Flatten())

model.add(Dense(120, activation='tanh'))
model.add(Dense(84, activation='tanh'))
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])
```

```python
callback = EarlyStopping(
    monitor = 'val_accuracy',
    min_delta = 0.00001,
    patience = 20,
    verbose = 1,
    baseline=None,
    restore_best_weights=True,
    mode = 'auto'
)
```

```python
model.fit(X_train_np,y_train, epochs = 500, validation_data =(X_test_np, y_test), batch_size=256, callbacks = callback)
```

```python
y_prob = model.predict(X_test_np)
```

```python
y_pred = y_prob.argmax(axis = 1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
